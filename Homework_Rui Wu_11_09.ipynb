You are given two tables - friend_request and request_accepted. Friend_request contains requester_id, time and sent_to_id and request_accepted table contains time, acceptor_id and requestor_id. How will you determine the overall acceptance rate of requests? (Write in both Python and SQL)

```python
friend_request = pd.DataFrame({
    'requester_id': ['501', '502', '503', '504'],
    'time': ['11/1/2017', '11/2/2017', '11/3/2017', '11/4/2017'],
    'send_to_id': ['801', '802', '803', '804']
})
```

```python
request_accepted = pd.DataFrame({
    'requester_id': ['501', '502', '503'],
    'acceptor_id': ['801', '802', '803']
})
```

```python
friend_request
```

```python
request_accepted
```

```python
data = pd.merge(friend_request, request_accepted, how='left', on='requester_id')
```

```python
data
```

```python
len(data['acceptor_id'][data.acceptor_id > 0]) / (len(data.requester_id)*1.0)
```

```python
card_number = ['4175004175004172'
,'4917491749174917'
,'346416800707698'
,'376416800707698'
,'37641680070769832112'
,'36641680070769'
,'54545641680070769']
```

```python
def issuer(card_number):
    if len(card_number)==16 and card_number[0]=='4':
        if (card_number[0:4]==['4026', '4405', '4508', '4844', '4913', '4917']) or (card_number[0:6]=='417500'):
            return 'VISAELECTRON'
        else:
            return 'VISA'
    elif len(card_number)==15 and (card_number[0:2]=='34' or card_number[0:2]=='37'):
        return 'AMEX'
    elif len(card_number)==14 and card_number[0:2]=='36':
        return 'Diners Club'
    else:
        return 'N/A'
```

```python
for card_number in card_number:
    print issuer(card_number)
```
